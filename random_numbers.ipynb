```python
#create 10 random floating point integers with a while loop in this code we use random.random to produce a random float. however as we need a range we first ask it to perform randint and supply the range
## the supplied range is then converted to a random float by adding the random.random function. we then send it through the while loop 10 times i.e. while x<11, which stops the while loop as we add to x

x=1
latitudes=[]

while x <11:
    random_lat=random.randint(-90,89)+ random.random()
    latitudes.append(random_lat)
    x += 1
random_lat

latitudes
```

```python
def latitudes(size):
    latitudes = []
    x = 0
    while x < (size):
        random_lat = random.randint(-90, 90) + random.random()
        latitudes.append(random_lat)
        x += 1
    return latitudes
# Call the function with 1500.
latitudes(1500)
```
